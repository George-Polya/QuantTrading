```python
import pandas_datareader as pdr
import pandas as pd

from datetime import datetime, timedelta
import matplotlib.pyplot as plt
```

```python
start_day = datetime(2002,4,16)
end_day = datetime.today()
end_day
```

```python
def get_price_data(x):
    df_price = pd.DataFrame(columns=x)

    for ticker in x:
        df_price[ticker] = pdr.get_data_yahoo(ticker, start_day - timedelta(days=365), end_day)["Adj Close"]

    return df_price
```

```python
stock = ["SSO"]
bond = ["TMF"]
df_stock = get_price_data(stock)
df_bond = get_price_data(bond)
```

# 월별 데이터 조회

```python
# 월별
df_stock_month = df_stock.resample(rule="1M").last()
df_bond_month = df_bond.resample(rule="1M").last()
```

```python
# 월별 수익율
df_stock_month.pct_change()
df_bond_month.pct_change()
```

```python
# 최근 12개월 월별 수익율
df_stock_12month = df_stock_month.pct_change().tail(12)
df_bond_12month = df_bond_month.pct_change().tail(12)
```

```python
stock_12month_std = df_stock_12month.std().item()
bond_12month_std = df_bond_12month.std().item()
print("SSO 최근 12개월 월별 변동성: ", stock_12month_std)
print("TMT 최근 12개월 월별 변동성: ",bond_12month_std)
```

```python
stock_weight = 1 - (stock_12month_std / (stock_12month_std+bond_12month_std))
bond_weight = 1 - stock_weight
```
